```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

start_date = datetime.strptime("2023-03-01", "%Y-%m-%d")
end_date = datetime.strptime("2023-05-01", "%Y-%m-%d")

tasks = {
    "": ("2023-03-01", 7),
    "": ("2023-03-08", 3),
    "": ("2023-03-11", 1),
    "": ("2023-03-12", 4),
    "": ("2023-03-16", 14),
    "": ("2023-03-30", 6),
    "": ("2023-04-05", 15),
    "": ("2023-04-20", 11)
}

task_dates = {task: (mdates.date2num(datetime.strptime(start, "%Y-%m-%d")), duration) 
              for task, (start, duration) in tasks.items()}

task_names = list(task_dates.keys())
task_starts = [dates[0] for dates in task_dates.values()]
task_lengths = [dates[1] for dates in task_dates.values()]

fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(task_names, task_lengths, left=task_starts, color='skyblue')
ax.set_title('Ганттова диаграмма разработки веб-сайта')
ax.set_xlabel('Даты')
ax.set_ylabel('Задачи')
ax.set_xlim([mdates.date2num(start_date), mdates.date2num(end_date)])

locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right")  

plt.gca().invert_yaxis()  
plt.grid(True)
plt.tight_layout()  
plt.show()
```
